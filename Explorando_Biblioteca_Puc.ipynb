```python
import pandas as pd

# Carregar o arquivo para verificar as colunas e os dados
file_path = '/content/Acervo MB.xlsx'
data = pd.ExcelFile(file_path)

# Verificar as planilhas disponíveis no arquivo
data.sheet_names
```

```python
# Carregar os dados da planilha
df = data.parse('NOVOS LIVROS - AGOSTO e SETEMBR')

# Exibir as primeiras linhas para análise das colunas
df.head()
```

```python
# Função para gerar termos de indexação
def gerar_termos_indexacao(row):
    termos = set()  # Usar um conjunto para evitar duplicatas

    # Adicionar palavras-chave se existirem
    if pd.notna(row['PALAVRAS-CHAVE MB']):
        termos.update(row['PALAVRAS-CHAVE MB'].split(','))

    # Adicionar área de conhecimento
    if pd.notna(row['ÁREA DO CONHECIMENTO MB']):
        termos.add(row['ÁREA DO CONHECIMENTO MB'])

    # Adicionar título como termo, limitando ao essencial
    if pd.notna(row['Título']):
        termos.add(" ".join(row['Título'].split()[:3]))  # Pegar as três primeiras palavras do título

    # Adicionar autoria como termo (primeiro autor)
    if pd.notna(row['AUTORIA']):
        termos.add(row['AUTORIA'].split(';')[0].strip())

    # Retornar até 5 termos, separados por vírgula
    return ', '.join(list(termos)[:5])

# Aplicar a função para preencher a coluna "Indexação"
df['Indexação'] = df.apply(gerar_termos_indexacao, axis=1)

# Visualizar o resultado para confirmar
df[['PALAVRAS-CHAVE MB', 'ÁREA DO CONHECIMENTO MB', 'Título', 'AUTORIA', 'Indexação']].head()
```

```python
# Salvar o DataFrame atualizado em um novo arquivo Excel
output_path = 'Acervo_MB_Indexado.xlsx'
df.to_excel(output_path, index=False)

output_path
```
